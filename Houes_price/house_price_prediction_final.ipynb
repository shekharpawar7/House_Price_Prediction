```python
#plotting floors and price data
plt.bar('floors','price', data = data_h, color = "blue")
plt.xlabel('floors')
plt.ylabel('price')
plt.show()
```

```python
# Selecting the features and target variable

Features1 = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode']
target = ['price']
```

```python
X1 = data_h[Features1]
```

```python
y1 = data_h[target]
```

```python
#importing linear regression model

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# splitting data into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.2, random_state=42)
```

```python
#fit linear_regression model

model = LinearRegression()
```

```python
# Training the model

model.fit(X_train, y_train)
```

```python
# Making predictions on the test set

y_pred = model.predict(X_test)
```

```python
# Evaluating the model

score = model.score(X_test, y_test)
print("Model R^2 Score:", score)
```

```python
# Predicting the price of a new house

new_house = pd.DataFrame({'bedrooms': [2], 'bathrooms': [2.5], 'sqft_living': [600], 'sqft_lot': [600], 'floors': [2], 'zipcode': [98008]})
predicted_price = model.predict(new_house)
print("Predicted Price:", predicted_price[0])
```
